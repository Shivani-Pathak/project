```python
# Importing the packages
import requests
from bs4 import BeautifulSoup
from textblob import TextBlob
import re
import nltk
from nltk.corpus import stopwords
import pandas as pd
import pyphen

# Initialize NLTK stopwords
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))

# Initialize Pyphen for syllable counting
dic = pyphen.Pyphen(lang='en_US')

# Define a function to calculate the number of syllables in a word
def count_syllables(word):
    try:
        return len(dic.inserted(word).split('-'))
    except ValueError:
        # If the word is too long, you can handle it as needed
        return None

# Define a function to calculate the FOG index
def calculate_fog_index(avg_sentence_length, percentage_complex_words):
    return 0.4 * (avg_sentence_length + percentage_complex_words)

# Define a function to perform text analysis
def analyze_text(url, text):
    # Sentiment Analysis using TextBlob
    sentiment = TextBlob(text)
    positive_score = sentiment.sentiment.polarity
    negative_score = -sentiment.sentiment.polarity
    polarity_score = sentiment.sentiment.polarity
    subjectivity_score = sentiment.sentiment.subjectivity

    # Tokenize the text into sentences and words
    sentences = nltk.sent_tokenize(text)
    words = nltk.word_tokenize(text)

    # Average Sentence Length
    avg_sentence_length = len(words) / len(sentences)

    # Percentage of Complex Words
    complex_word_count = sum(1 for word in words if len(word) > 6)  # Define your complexity criteria
    percentage_complex_words = (complex_word_count / len(words)) * 100

    # Calculate the number of syllables for each word in the text
    syllable_count = [count_syllables(word) for word in words]

    # Handle words with too many syllables as needed
    for i, syllables in enumerate(syllable_count):
        if syllables is None:
            syllable_count[i] = 0  # Set a default value for long words

    # Calculate the total syllables in the text
    total_syllables = sum(syllable_count)

    # Personal Pronouns (You can use regular expressions)
    personal_pronouns = len(re.findall(r'\b(I|me|my|mine|myself|we|us|our|ours|ourselves|you|your|yours|yourself|yourselves|he|him|his|himself|she|her|hers|herself|it|its|itself|they|them|their|theirs|themselves)\b', text, re.IGNORECASE))

    # Average Word Length
    avg_word_length = sum(len(word) for word in words) / len(words)

    # Calculate the FOG index
    fog_index = calculate_fog_index(avg_sentence_length, percentage_complex_words)

    # Average Number of Words per Sentence
    avg_words_per_sentence = len(words) / len(sentences)

    # Word Count
    word_count = len(words)

    # Return the computed variables as a dictionary
    print("URL:", url)
    print("Positive Score:", positive_score)
    print("Negative Score:", negative_score)
    print("Polarity Score:", polarity_score)
    print("Subjectivity Score:", subjectivity_score)
    print("Average Sentence Length:", avg_sentence_length)
    print("Percentage of Complex Words:", percentage_complex_words)
    print("FOG Index:", fog_index)
    print("Average Number of Words per Sentence:", avg_words_per_sentence)
    print("Complex Word Count:", complex_word_count)
    print("Word Count:", word_count)
    print("Personal Pronouns:", personal_pronouns)
    print("Average Word Length:", avg_word_length)



# Read URLs from an Excel file
path = '/content/MyDrive/MyDrive/Input.xlsx'
data = pd.read_excel(path)
urls = data["URL"]


# Loop through the URLs and perform text analysis
results = []
for url in urls:
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        text = soup.get_text()
        analysis_result = analyze_text(url, text)
        results.append(analysis_result)
    else:
        print(f"Failed to fetch URL: {url}")

# Create a DataFrame to store the results
df = pd.DataFrame(results)

# Save the results to a CSV file
df.to_csv("text_analysis_results.csv", index=False)

```
